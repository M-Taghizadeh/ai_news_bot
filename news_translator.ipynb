# Translator

```python
import pandas as pd

df = pd.read_csv("ai_news_bot/news_content.csv")
```

```python
import googletrans

news_dict = {'title': [], 'title_fa': [], 'url': [], 'content': [], 'content_summary': [], 'content_fa': [], 'thumbnail': [], 'publish_date': []}
number_of_news = len(df['title'])
number_of_news = 10

for i in range(0, number_of_news):
    row = df.iloc[i]
    title = row['title']
    content_summary = row['content_summary']

    try:
        # translate content
        translator = googletrans.Translator()
        title_fa = translator.translate(title, src='en', dest='fa')
        content_fa = translator.translate(content_summary, src='en', dest='fa')

        news_dict['title'].append(title)
        news_dict['title_fa'].append(title_fa.text)
        news_dict['url'].append(row['url'])
        news_dict['content'].append(row['content'])
        news_dict['content_summary'].append(row['content_summary'])
        news_dict['content_fa'].append(content_fa.text)
        news_dict['thumbnail'].append(row['thumbnail'])
        news_dict['publish_date'].append(row['publish_date'])
    except: print("error")
```

```python
news_df = pd.DataFrame(news_dict)
news_df
```
